```python
df = pd.read_csv("Path to data without anomalies", sep=', ', delimiter=',', header='infer')
dfdata1 = pd.DataFrame(df)  
```

```python
pd.set_option('max_rows',20)
pd.set_option('max_columns',50)
```

```python
dfdata = dfdata1.iloc[0:30000,:] 
dfdata
```

```python
np.random.seed(4)
Attack_or_Not = np.random.uniform(low=0.0, high=1.0, size=30000) 
Attack_or_Not[0:10]
```

```python
In_vehicle = dfdata['InVehicle_Longitudinal_Speed']
In_vehicle.values
```

```python
GPS_Speed = dfdata['GPS_Speed']
```

```python
Acc = dfdata['InVehicle_Longitudinal_Accel']
```

```python
np.random.seed(4)
dur = 10
class_output = list()
for i in range(29800):  
    if (Attack_or_Not[i] < 0.05) & (Attack_or_Not[i] >= 0.033):
        Magn = round(np.random.uniform(0,5),3) 
        In_vehicle[i:i+dur] = In_vehicle.values[i] + Magn
        
    elif (Attack_or_Not[i] < 0.033) & (Attack_or_Not[i] >= 0.0167):
        Magn = round(np.random.uniform(0,5),3)
        GPS_Speed[i:i+dur] = GPS_Speed.values[i] + Magn
        
    elif (Attack_or_Not[i] < 0.0167):
        Magn = round(np.random.uniform(0,5),3) 
        Acc[i:i+dur] = Acc.values[i] + Magn
        
```

```python
class_output_K = list()
for i in range (0,29800):  #9800   
    if ((Attack_or_Not[i] < 0.05) & (Attack_or_Not[i] >= 0.033))| ((Attack_or_Not[i-1] < 0.05) & (Attack_or_Not[i-1] >= 0.033))| ((Attack_or_Not[i-2] < 0.05) &  (Attack_or_Not[i-2] >= 0.033))|\
    ((Attack_or_Not[i-3] < 0.05)& (Attack_or_Not[i-3] >= 0.033))| ((Attack_or_Not[i-4] < 0.05)& (Attack_or_Not[i-4] >= 0.033))| ((Attack_or_Not[i-5] < 0.05) & (Attack_or_Not[i-5] >= 0.033))| ((Attack_or_Not[i-6] < 0.05)& (Attack_or_Not[i-6] >= 0.033))|\
    ((Attack_or_Not[i-7] < 0.05) & (Attack_or_Not[i-7] >= 0.033))| ((Attack_or_Not[i-8] < 0.05)& (Attack_or_Not[i-8] >= 0.033))| ((Attack_or_Not[i-9] < 0.05)& (Attack_or_Not[i-9] >= 0.033)):
        class_output_K.append(1)                              
 
    else:
        class_output_K.append(0)      
    
    
print(class_output_K)
```
